### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg=log_reg.fit(x_train,y_train)
```

```python
y_pred_LogReg=log_reg.predict(x_test)
```

```python
from sklearn.metrics import f1_score,accuracy_score
```

```python
print('f1_score using logistic regression is :',f1_score(y_test,y_pred_LogReg))
```

### k-NN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k=[]
neighbors=np.arange(3,42)
```

```python
for k in neighbors:
    kNN_Class=KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    kNN_Class.fit(x_train,y_train)
    y_pred_kNN=kNN_Class.predict(x_test)
    f1sc=f1_score(y_test,y_pred_kNN)
    metric_k.append(f1sc)
    
```

```python
metric_k
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k-value')
plt.ylabel('F1_score')
plt.grid()
```

```python
kNN_Class=KNeighborsClassifier(n_neighbors=2, metric='euclidean')
kNN_Class.fit(x_train,y_train)
y_pred_kNN=kNN_Class.predict(x_test)
```

```python
f1_score(y_test,y_pred_kNN)
```

```python
kNN_Class=KNeighborsClassifier(n_neighbors=3, metric='euclidean')
kNN_Class.fit(x_train,y_train)
y_pred_kNN1=kNN_Class.predict(x_test)
```

```python
f1_score(y_test,y_pred_kNN1)
```

```python
# model is better when  number of neighbors=3
```
